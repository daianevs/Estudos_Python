### 📘 EXERCÍCIO – Cálculo de Empréstimo Consignado com Desconto

Um cliente está simulando um empréstimo consignado com duas opções de prazo: até 12 meses ou acima de 12 meses.  
A taxa de juros muda de acordo com o prazo escolhido.

Vamos calcular o valor total do empréstimo com base nas regras abaixo:

---

🔁 **Regras:**

- **Prazo até 12 meses:** Taxa de juros ao mês: **1,5%**
- **Prazo acima de 12 meses:** Taxa de juros ao mês: **1,2%**

💡 **Desconto adicional:**  
Se o valor solicitado for **acima de R$ 10.000** ou o **prazo maior que 24 meses**, o cliente recebe um **desconto de 5%** sobre o valor total final (como incentivo promocional).

---

🧠 **Tarefa:**  
Escreva um código que:

1. Solicite ao cliente o valor do empréstimo desejado.
2. Solicite o número de parcelas (prazo em meses).
3. Calcule o valor final com base na taxa de juros correspondente.
4. Verifique se há direito ao desconto.
5. Exiba um **resumo da simulação** como o exemplo abaixo:

---

### __📊 SIMULAÇÃO DE EMPRÉSTIMO__

```
Valor solicitado           : R$ 12.000,00  
Prazo                      : 30 meses  
Taxa de Juros aplicada     : 1,2%  
Valor final com juros      : R$ 15.203,45  
Desconto aplicado          : R$ 760,17  
Valor total com desconto   : R$ 14.443,28
```


## Informações cadastrais do crédito

````python
# Solicita ao usuário o valor do empréstimo e a quantidade de parcelas (prazo em meses)

consignado = float(input('Digíte o valor do Empréstimo Contratado: '))
parcelas = int(input('Digíte o número de parcelas: '))

# Define a taxa de juros com base no número de parcelas
# - Se o prazo for até 12 meses: aplica 1,5%
# - Acima de 12 meses: aplica 1,2%
# A variável taxa_juros armazena a versão decimal (ex: 1,5% → 0.015) usada nos cálculos

taxa_juros = 0
if parcelas <= 12:
  taxa_j = 1.5
  taxa_juros = 0.015
else:
  taxa_j = 1.2
  taxa_juros = 0.012


````

### 📐 Cálculo de Matemática Financeira

Utilizamos a fórmula dos **juros compostos** para simular a evolução do valor do empréstimo ao longo do tempo.  
Essa fórmula permite calcular o **montante final a ser pago**, levando em consideração a taxa de juros mensal e o número de parcelas contratadas.



````python
# Aplica a fórmula de juros compostos: M = C × (1 + j)^t
# Onde:
# M = Montante (valor final com juros)
# C = Capital (valor solicitado)
# j = Taxa de juros mensal (em decimal)
# t = Prazo (número de parcelas)


valor_final_consignado = consignado * (1 + taxa_juros)** parcelas

# Verifica direito ao desconto promocional
# Inicializa a variável com zero para evitar erro de referência
# O desconto de 5% é aplicado se:
# - O valor solicitado for maior que 10.000 (sem ponto)
# - OU o prazo for superior a 24 parcelas

desconto_consignado = 0

if consignado > 10000 or parcelas > 24:
  desconto_consignado = valor_final_consignado * 0.05

valor_final_consig_desconto = valor_final_consignado - desconto_consignado

# Cálculo do valor da parcela mensal
# Inicializa a variável com zero para evitar erro de referência
# Existem duas possibilidades:
# - Se houver desconto, calcula com o valor final com desconto
# - Caso contrário, calcula com o valor total com juros


valor_parcela = 0
if desconto_consignado > 0:
  valor_parcela = valor_final_consig_desconto / parcelas
else:
  valor_parcela = valor_final_consignado / parcelas
````

## Área dos prints

````python
print('\n________ SIMULAÇÃO CONSIGNADO________\n')

print('Valor Solicitado                : R${:,.2f}'.format(consignado))
print(f'Prazo                           : {parcelas}')
print(f'Taxa de Juros Aplicada          : {taxa_j}%')
print('Valor final com Juros           : R${:,.2f}'.format(valor_final_consignado))
print('Desconto Aplicado               : R${:,.2f}'.format(desconto_consignado))
print('Valor Total com o Desconto      : R${:,.2f}'.format(valor_final_consig_desconto))
print(f'Valor da parcela                : R$ {valor_parcela:,.2f}')

### Observação sobre formatação numérica:
# {:,.2f} → Exibe números com separador de milhar e duas casas decimais
# - A vírgula (,) separa os milhares → ex: 10000 → 10,000.00
# - O .2f limita o número a duas casas decimais → ex: 10.456 → 10.46
````
